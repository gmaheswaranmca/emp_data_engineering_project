````python

# ðŸ“Š Sales Dashboard Project using Pandas and Matplotlib

## Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt


````

````python
import io


## Step 2: Sample Data Creation
'''
data = {
    'Date': pd.date_range(start='2023-01-01', periods=12, freq='ME'),
    'Sales': [12000, 15000, 17000, 13000, 19000, 21000, 18000, 20000, 22000, 25000, 24000, 26000],
    'Profit': [3000, 4000, 4500, 3200, 4700, 5000, 4600, 4800, 5300, 6000, 5900, 6100],
    'Region': ['North', 'East', 'West', 'South', 'North', 'East', 'West', 'South', 'North', 'East', 'West', 'South']
}'''

csv_text = """Date,Sales,Profit,Region
2023-01-31,12000,3000,North
2023-02-28,15000,4000,East
2023-03-31,17000,4500,West
2023-04-30,13000,3200,South
2023-05-31,19000,4700,North
2023-06-30,21000,5000,East
2023-07-31,18000,4600,West
2023-08-31,20000,4800,South
2023-09-30,22000,5300,North
2023-10-31,25000,6000,East
2023-11-30,24000,5900,West
2023-12-31,26000,6100,South
"""

df = pd.read_csv(io.StringIO(csv_text), parse_dates=['Date'])
#df = pd.DataFrame(data)
````

````python

## Step 3: Line Plot - Sales Over Time
df.plot(x='Date', y='Sales', kind='line', marker='o', title='Monthly Sales')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

````

````python

## Step 4: Line Plot - Profit Over Time
df.plot(x='Date', y='Profit', kind='line', marker='s', color='green', title='Monthly Profit')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
````

````python

## Step 5: Bar Plot - Sales by Region
df.groupby('Region')['Sales'].sum().plot(kind='bar', color='skyblue', title='Total Sales by Region')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.show()

````

````python

## Step 6: Pie Chart - Sales by Region

df.groupby('Region')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Sales Share by Region')
plt.ylabel('')
plt.show()

````

````python

## Step 7: Save Plot

df.plot(x='Date', y='Sales', kind='line')
plt.title('Monthly Sales')
plt.savefig('monthly_sales_plot.png', dpi=300, bbox_inches='tight')
plt.show()
````
